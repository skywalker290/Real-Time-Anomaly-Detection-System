```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('d1_anomaly_detection_dataset.csv')

# Extract the category (train/Normal, train/Anomaly, etc.) from the Path column
df['Category'] = df['Path to Video'].apply(lambda x: '/'.join(x.split('/')[-3:-1]))

# Categorize as 'Normal' or 'Anomaly' for coloring
df['Type'] = df['Category'].apply(lambda x: 'Normal' if 'Normal' in x else 'Anomaly')

# Count the number of videos in each category
category_counts = df.groupby(['Category', 'Type']).size().reset_index(name='Count')

# Plotting
colors = {'Normal': 'green', 'Anomaly': 'red'}

plt.figure(figsize=(10, 6))

# Bar chart
for category_type in ['Normal', 'Anomaly']:
    subset = category_counts[category_counts['Type'] == category_type]
    plt.bar(
        subset['Category'],
        subset['Count'],
        color=colors[category_type],
        label=category_type,
        edgecolor='black'
    )

# Adding titles and labels
plt.title('Count of Videos by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Videos', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Video Type', fontsize=12)
plt.tight_layout()

# Show the plot


output_file = 'video_counts_by_category.png'
plt.savefig(output_file, bbox_inches='tight')  # Adjust DPI and layout for high quality
print(f"Plot saved as {output_file}")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the new CSV file
df = pd.read_csv('d2_anomaly_muticlass_dataset.csv')

# Extract the category (train/Robbery, train/Abuse, etc.) from the Path column
df['Category'] = df['Path to Video'].apply(lambda x: '/'.join(x.split('/')[-3:-1]))

# Extract the subset (train, test, val) from the Path column
df['Subset'] = df['Path to Video'].apply(lambda x: x.split('/')[2])

# Count the number of videos for each category and subset
category_subset_counts = df.groupby(['Category', 'Subset']).size().unstack(fill_value=0)

# Sort the categories to ensure the categories appear in the correct order
category_subset_counts = category_subset_counts.sort_index()

# Create a grouped bar chart
# Plotting with default figure size
category_subset_counts = category_subset_counts.reset_index()

# Set position for each group of bars
bar_width = 0.25
category_subset_counts.set_index('Category', inplace=True)
categories = category_subset_counts.index
num_categories = len(categories)

# Create figure
plt.figure(figsize=(16, 8))

# Generate the bars for train, test, val
bar1 = plt.barh(np.arange(num_categories) - bar_width, category_subset_counts['train'], bar_width, label='Train', edgecolor='black')
bar2 = plt.barh(np.arange(num_categories), category_subset_counts['test'], bar_width, label='Test', edgecolor='black')
bar3 = plt.barh(np.arange(num_categories) + bar_width, category_subset_counts['val'], bar_width, label='Val', edgecolor='black')

# Adding titles and labels
plt.title('Count of Videos by Category and Subset')
plt.xlabel('Number of Videos')
plt.ylabel('Category')

# Set y-ticks to the middle of each group
plt.yticks(np.arange(num_categories), categories)

# Add legend
plt.legend(title="Subset", loc='upper right')

# Save the plot as an image
output_file = 'grouped_video_counts_by_category.png'
plt.tight_layout()  # Ensure everything fits
plt.savefig(output_file, dpi=300, bbox_inches='tight')  # Save with high resolution and tight bounding box
print(f"Plot saved as {output_file}")

# Show the plot
plt.show()
```
